```python
import pandas as pd
```

```python
trips = pd.read_csv('../data/trentino_trasporti/trips.txt')
```

```python
routes = pd.read_csv('../data/trentino_trasporti/routes.txt')
```

```python
shapes = pd.read_csv('../data/trentino_trasporti/shapes.txt')
```

```python
stop_times = pd.read_csv('../data/trentino_trasporti/stop_times.txt')
```

merge trips with routes by route_id so that each trip has a route associated to it now and a shape as well

```python
final_df = trips.merge(routes, on='route_id').merge(shapes, on='shape_id').merge(stop_times, on='trip_id')
```

fromt this dataset we have information about each trip that should be carried out each day including also a 

```python
final_df = final_df.drop(['route_color', 'route_text_color', 'route_type', 'agency_id', 'service_id'], axis=1)
```
