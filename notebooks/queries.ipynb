```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Connection settings
db_user = 'postgres'
db_pass = 'example'
db_host = 'localhost'
db_port = '5432'
db_name = 'raw_data'

# SQLAlchemy connection string
engine = create_engine(f'postgresql+psycopg2://{db_user}:{db_pass}@{db_host}:{db_port}/{db_name}')
```

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("SELECT MAX(bus_id) FROM bus"))
    for row in result:
        print(row[0])
```

```python
# Query the database
df = pd.read_sql("SELECT stop_times.stop_id, routes.route_short_name, stop_times.arrival_time, stop_times.departure_time, stop_times.trip_id, stop_times.stop_sequence FROM trips JOIN stop_times ON trips.trip_id = stop_times.trip_id JOIN routes ON trips.route_id = routes.route_id", engine)
```
